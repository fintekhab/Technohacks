**Calculating Mean, Median and Mode of the Fare of the Passengers**

```python
d['Fare'].mean()
```

```python
d['Fare'].median()
```

```python
d['Fare'].mode()
```

```python
plt.figure(figsize=(8,6))
plt.hist(d['Fare'], color='maroon')
plt.axvline(d['Fare'].mean(), color='green', linestyle='dashed', linewidth=1.5, label=f"Mean: {d['Fare'].mean():.2f}")
plt.axvline(d['Fare'].median(), color='orange', linestyle='dashed', linewidth=1.5, label=f"Meadian: {d['Fare'].median():.2f}")
fare_modes = d['Fare'].mode()
for mode_value in fare_modes:
  plt.axvline(mode_value, color='blue', linestyle = 'dashed', linewidth=1.5, label=f"Mode: {mode_value:.2f}")
plt.xlabel('Fare', color='red')
plt.ylabel('Frequency', color='red')
plt.title('Distribution of Fare with Mean, Median, and Mode', color='red')
plt.legend()
plt.show()
```
